```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Define the figure names
first_row = [
    "Training_Predictions_Target_1P1.png",
    "Training_Predictions_Target_1P2.png",
    "Training_Predictions_Target_1P3.png",
    "Training_Predictions_Target_1P4.png",
    "Training_Predictions_Target_1P5.png"
]

second_row = [
    "Training_Predictions_Target_2P1.png",
    "Training_Predictions_Target_2P2.png",
    "Training_Predictions_Target_2P3.png",
    "Training_Predictions_Target_2P4.png",
    "Training_Predictions_Target_2P5.png"
]

# Load the images
images = [mpimg.imread(img) for img in first_row + second_row]

# Create a figure
fig, axes = plt.subplots(2, 5, figsize=(20, 8))

# Plot images in a grid
for i, ax in enumerate(axes.flat):
    ax.imshow(images[i])
    ax.axis('off')  # Hide the axes

# Adjust layout
plt.subplots_adjust(wspace=0, hspace=0)

# Save or show the combined figure
plt.savefig('combined_figure.png')
# Save as a high-quality PDF
plt.savefig('combined_figure.pdf', format='pdf', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Define the figure names
first_column = [
    "Training_PredictionsTarget1_P1.png",
    "Training_Predictions_Target_1_P2.png",
    "Training_Predictions_Target_1_P3.png",
    "Training_Predictions_Target_1_P4.png",
    "Training_Predictions_Target_1_P5.png"
]
second_column = [
    "Training_PredictionsTarget2_P1.png",
    "Training_Predictions_Target_2_P2.png",
    "Training_Predictions_Target_2_P3.png",
    "Training_Predictions_Target_2_P4.png",
    "Training_Predictions_Target_2_P5.png"
]

# Load the images
images = [mpimg.imread(img) for img in first_column + second_column]

# Create a figure
fig, axes = plt.subplots(5, 2, figsize=(10, 20))  # 5 rows, 2 columns

# Plot images in a grid
for i, ax in enumerate(axes.flat):
    ax.imshow(images[i])
    ax.axis('off')  # Hide the axes
    
    # Add labels to the top left of each figure, adjusted position
    label = f"({chr(97 + i)})"  # This will create labels a, b, c, ...
    ax.text(0, 1.1, label, transform=ax.transAxes, fontsize=12, fontweight='bold', va='top', ha='left')

# Adjust layout
plt.tight_layout()
plt.subplots_adjust(wspace=0.01, hspace=0.05)  # Reduced wspace to bring columns closer

# Save or show the combined figure
plt.savefig('Fig 10.png', bbox_inches='tight', pad_inches=0.1)
# Save as a high-quality PDF
plt.savefig('Fig 10.pdf', format='pdf', dpi=300, bbox_inches='tight', pad_inches=0.1)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Define the figure names
figure_names = [
    "MSEvsEpochs_P1.png",
    "MSEvsEpochs_P2.png",
    "MSEvsEpochs_P3.png",
    "MSEvsEpochs_P4.png",
    "MSEvsEpochs_P5.png"
]

# Define the labels
labels = ['a', 'b', 'c', 'd', 'e']

# Load the images
images = [mpimg.imread(img) for img in figure_names]

# Create a figure with a 3x3 grid
fig, axes = plt.subplots(3, 3, figsize=(20, 24))

# Plot images in the desired arrangement
plot_positions = [(0, 0), (0, 2), (1, 0), (1, 2), (2, 1)]

for pos, img, label in zip(plot_positions, images, labels):
    ax = axes[pos[0], pos[1]]
    ax.imshow(img)
    title_config = {
        'fontsize': 20,
        'fontweight': 'bold',  # Make text bold
        'loc': 'left'
    }
    if label == 'e':  # Check if it's the label 'e'
        title_config['pad'] = 40  # Increase padding for label 'e'
    ax.set_title(f'({label})', **title_config)
    ax.axis('off')

# Remove the extra subplots
for i, j in [(0, 1), (1, 1), (2, 0), (2, 2)]:
    fig.delaxes(axes[i, j])

# Hide the axes for all remaining subplots
for ax in axes.flat:
    if ax.get_visible():
        ax.axis('off')

# Adjust the spacing between subplots to bring them very close
plt.subplots_adjust(wspace=-0.45, hspace=+0.15)

# Save or show the combined figure
plt.savefig('Fig 12.png', bbox_inches='tight', pad_inches=0)
# Save as a high-quality PDF
plt.savefig('Fig 12.pdf', format='pdf', dpi=300, bbox_inches='tight', pad_inches=0)
plt.show()
```
